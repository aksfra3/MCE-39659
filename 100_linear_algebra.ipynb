```julia
using LinearAlgebra

# 1. Input the data from the screenshot
T = 15                   # Number of observations
k = 3                    # Number of parameters (beta1, beta2, beta3)
RSS = 10.96              # Residual Sum of Squares (u'u)

# (X'X)^-1 matrix
inv_XtX = [ 2.0  3.5 -1.0;
            3.5  1.0  6.5;
           -1.0  6.5  4.3]

# (X'y) vector
Xty = [-3.0, 2.2, 0.6]

# 2. Calculate Coefficient Estimates (beta_hat)
# Formula: beta_hat = (X'X)^-1 * (X'y)
beta_hat = inv_XtX * Xty

display(inv_XtX)
display(Xty)
display(beta_hat)

```

```julia
#estimate of variance 

var_beta_hat = RSS / (T - k)

display(var_beta_hat)
```

```julia
#estimate variance covariance matrix

varcov_beta_hat = var_beta_hat * inv_XtX
display(varcov_beta_hat)
```

```julia
#estimate standard Errors

standard_errors = sqrt.(diag(varcov_beta_hat))
display(standard_errors)
```

```julia
gg1 = sqrt(varcov_beta_hat[1,1])
gg2 = sqrt(varcov_beta_hat[2,2])

println(gg1,"               ",gg2)
```

```julia
# 3. Calculate the Variance of the residuals (sigma_squared_hat)
# Formula: sigma^2 = RSS / (T - k)
sigma_sq_hat = RSS / (T - k)

# 4. Calculate the Variance-Covariance Matrix of the coefficients
# Formula: Var(beta_hat) = sigma^2 * (X'X)^-1
var_beta_hat = sigma_sq_hat * inv_XtX

# 5. Calculate Standard Errors
# Standard error is the square root of the diagonal elements of the Var-Cov matrix
standard_errors = sqrt.(diag(var_beta_hat))

# --- Display Results ---
println("Coefficient Estimates (beta_hat):")
for i in 1:k
    println("  beta_$i: ", round(beta_hat[i], digits=4))
end

println("\nStandard Errors:")
for i in 1:k
    println("  SE(beta_$i): ", round(standard_errors[i], digits=4))
end
```
