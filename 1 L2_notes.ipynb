## logical operators


```julia
function test(x)
    y = 10
    z = x + 10
    return z
end

x = ones(5)
test(x)

```

```julia
function test(x)
    y = 10
    z = x       #integer
    z = z.+ y   #vector/array
    return z
end

x = ones(5)
# test(x)

# x = 6
test(x)
@time test(x)
```

```julia
function test(x::Float64)
    y = 10
    z = x       #integer
    z = z.+ y   #vector/array
    return z
end


x = ones(5)
# test(x)

# x = 6
test(x)
@time test(x)
```
