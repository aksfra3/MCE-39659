```julia
#Simple Newton Method for one variable (LOOP)

f(x) = x^2 - 2
f_prime(x) = 2x
f_double_prime(x) = 2

x = 1
tol = 1e-5
iter = 100

for i in 1:iter

    x_new = x - (f_prime(x) / f_double_prime(x))
    println("Iteration $i: x = $x, f(x) = $(f(x))")

    if abs(x_new - x) < tol
        println("Converged at iteration $i. Final x: $x, f(x) = $(f(x))")
        break
    end

    x = x_new

end
```

## Minimize Package "Optim"

```julia
#Package for optimizatino

import Pkg; Pkg.add("Optim")
using Optim
```

```julia
#Two variables

f(x) = (x[1] - 5.0)^2 + (x[2] - 2.0)^2

x0 = [0.0, 0.0]

result = optimize(f, x0)

println("Minimizer: ", Optim.minimizer(result))
println("Minimum: ", Optim.minimum(result))
```

```julia
#One variable 

f(x) = x^2 - 2

result = optimize(f, -10, 10)

println(Optim.minimizer(result), "                      " ,Optim.minimum(result))
```
