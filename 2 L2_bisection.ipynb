## Method: Bisecton
1) Guess
2) Mid point
3) chech f(a) * f(b) > 0
4) Find new x value (if function)
5) Check convergence

### Bisection 1
- while loop

```julia
function bisection1(f, a, b, tol = 1e-5)

    if f(a) *f(b) > 0 
        print("Root might not be in intevral")
        return nothing

    end

    while (b - a) / 2 > tol
        c = (a + b) / 2

        if f(c) == 0
            return c 
        
        elseif f(a) * f(c) < 0 
            b = c 
        else 
            a = c
        end

    end

    return (a + b) / 2
end

f(x) = x^2 - 2

root = bisection1(f, 1, 4)

print("The root is: ", root)
```

## Bisection 2
- for loop

```julia
function bisection2(f, a, b, N)

    tol = 1e-4
    
    for i in 1:N

        x = (a+b) / 2
        error1 = (b-a)/2

        println("i: ",i,"       ", "x:  ", round(x, digits = 5),"        ", "ϵ: ", round(error1, digits = 5))

        if f(a) * f(b) > 0 
            print("Root not in interval")
            return nothing
        end


        if abs(x - a) <= tol
            print("The root is:     ")
            return round(x, digits = 5)


        elseif f(a) * f(x) < 0
            b = x
        else 
            a = x
        end

    
    end

    return x
end



f(x) = x^2 - 2
root = bisection2(f, 1, 4, 100)


print(root)



    
```
